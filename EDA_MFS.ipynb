```python
# MFS Growth Prediction - Comprehensive Exploratory Data Analysis
# For Google Colab Environment
# Fixed Version with Proper Units and Complete Analysis

# Install required packages
!pip install plotly seaborn pandas matplotlib numpy scikit-learn

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy import stats
from scipy.stats import pearsonr

warnings.filterwarnings('ignore')

# Set styling
plt.style.use('default')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("üìä MFS Growth Prediction - Exploratory Data Analysis")
print("=" * 60)

# ================================
# PHASE 1: DATA LOADING & OVERVIEW
# ================================

# Upload your Excel file in Colab
from google.colab import files
print("Please upload your MFS Excel file:")
uploaded = files.upload()

# Load the dataset
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

print(f"\nüîç Dataset Overview:")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"Date range: {df['Months'].iloc[0]} to {df['Months'].iloc[-1]}")

# Basic info
print(f"\nüìã Dataset Info:")
df.info()

# Statistical summary
print(f"\nüìä Statistical Summary:")
print(df.describe())

# Missing values analysis
print(f"\n‚ùå Missing Values:")
missing_vals = df.isnull().sum()
missing_percent = (missing_vals / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_vals,
    'Missing_Percent': missing_percent
}).sort_values('Missing_Count', ascending=False)
missing_summary = missing_df[missing_df['Missing_Count'] > 0]
if len(missing_summary) > 0:
    print(missing_summary)
else:
    print("‚úÖ No missing values found!")

# ================================
# PHASE 2: DATA PREPROCESSING
# ================================

print(f"\nüîß DATA PREPROCESSING")
print("=" * 40)

# Create proper datetime column
df['Date'] = pd.to_datetime(df['Months'].str.replace("'", "20"), format='%B%Y', errors='coerce')
df = df.sort_values('Date').reset_index(drop=True)

# Define variable groups
transaction_counts = [col for col in df.columns if '_count' in col and col != 'Total_trans_count']
transaction_amounts = [col for col in df.columns if '_amount' in col and col != 'Total_amount' and col != 'Float_amount']
key_metrics = ['Total_trans_count', 'Total_amount', 'Float_amount']

print(f"\nüè∑Ô∏è Variable Groups:")
print(f"Transaction Counts ({len(transaction_counts)}): {transaction_counts}")
print(f"Transaction Amounts ({len(transaction_amounts)}): {transaction_amounts}")
print(f"Key Metrics ({len(key_metrics)}): {key_metrics}")

# Data units explanation
print(f"\nüìä Data Units Understanding:")
print("‚Ä¢ Transaction Counts: Actual number of transactions")
print("‚Ä¢ Transaction Amounts: In millions of BDT")
print("‚Ä¢ Float Amount: In millions of BDT")

# ================================
# PHASE 3: TEMPORAL ANALYSIS
# ================================

print(f"\nüìà TEMPORAL ANALYSIS")
print("=" * 40)

# Create time series plots for key metrics
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Total Transaction Count', 'Total Transaction Amount (Million BDT)',
                   'Float Amount (Million BDT)', 'Monthly Growth Rate'),
    specs=[[{"secondary_y": False}, {"secondary_y": False}],
           [{"secondary_y": False}, {"secondary_y": False}]]
)

# Total Transaction Count
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Total_trans_count'],
              mode='lines+markers', name='Transaction Count',
              line=dict(color='blue', width=2)),
    row=1, col=1
)

# Total Transaction Amount
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Total_amount'],
              mode='lines+markers', name='Transaction Amount',
              line=dict(color='green', width=2)),
    row=1, col=2
)

# Float Amount
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Float_amount'],
              mode='lines+markers', name='Float Amount',
              line=dict(color='red', width=2)),
    row=2, col=1
)

# Growth rates
df['Total_trans_count_growth'] = df['Total_trans_count'].pct_change() * 100
fig.add_trace(
    go.Scatter(x=df['Date'][1:], y=df['Total_trans_count_growth'][1:],
              mode='lines+markers', name='Transaction Count Growth (%)',
              line=dict(color='purple', width=2)),
    row=2, col=2
)

fig.update_layout(height=800, title_text="Key MFS Metrics Over Time", showlegend=False)
fig.show()

# Calculate growth statistics
print(f"\nüìà Growth Statistics:")
for metric in key_metrics:
    start_val = df[metric].iloc[0]
    end_val = df[metric].iloc[-1]
    total_growth = ((end_val - start_val) / start_val) * 100
    periods = len(df) - 1
    cagr = ((end_val / start_val) ** (1/periods) - 1) * 100

    unit = "Million BDT" if "amount" in metric.lower() else "Transactions"
    print(f"\n{metric}:")
    print(f"  Start: {start_val:,.0f} {unit}")
    print(f"  End: {end_val:,.0f} {unit}")
    print(f"  Total Growth: {total_growth:.1f}%")
    print(f"  Monthly CAGR: {cagr:.2f}%")

# ================================
# PHASE 4: TRANSACTION TYPE ANALYSIS
# ================================

print(f"\nüí≥ TRANSACTION TYPE ANALYSIS")
print("=" * 40)

# Transaction type evolution
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Transaction counts by type
transaction_count_cols = ['CashIn_count', 'CashOut_count', 'P2P_count', 'MP_count', 'G2P_count']
df[transaction_count_cols].plot(ax=axes[0,0], title='Transaction Counts by Type Over Time')
axes[0,0].set_xlabel('Month Index')
axes[0,0].set_ylabel('Transaction Count (Millions)')
axes[0,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0,0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))

# Transaction amounts by type
transaction_amount_cols = ['CashIn_amount', 'CashOut_amount', 'P2P_amount', 'MP_amount', 'G2P_amount']
df[transaction_amount_cols].plot(ax=axes[0,1], title='Transaction Amounts by Type Over Time')
axes[0,1].set_xlabel('Month Index')
axes[0,1].set_ylabel('Transaction Amount (Million BDT)')
axes[0,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Market share evolution (counts)
df_counts_pct = df[transaction_count_cols].div(df[transaction_count_cols].sum(axis=1), axis=0) * 100
df_counts_pct.plot(kind='area', stacked=True, ax=axes[1,0], title='Market Share - Transaction Counts (%)')
axes[1,0].set_xlabel('Month Index')
axes[1,0].set_ylabel('Percentage')
axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Market share evolution (amounts)
df_amounts_pct = df[transaction_amount_cols].div(df[transaction_amount_cols].sum(axis=1), axis=0) * 100
df_amounts_pct.plot(kind='area', stacked=True, ax=axes[1,1], title='Market Share - Transaction Amounts (%)')
axes[1,1].set_xlabel('Month Index')
axes[1,1].set_ylabel('Percentage')
axes[1,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# Fixed Average Transaction Values Analysis
print(f"\nüí∞ Average Transaction Values by Type:")
print("Note: Transaction amounts are in millions of BDT, so we multiply by 1,000,000 to get BDT per transaction")
print("-" * 80)

avg_transaction_values = {}
for i, count_col in enumerate(transaction_count_cols):
    amount_col = transaction_amount_cols[i]
    # Calculate average transaction value in BDT (amounts are in millions)
    mask = df[count_col] > 0
    if mask.sum() > 0:
        avg_val_bdt = (df.loc[mask, amount_col] * 1_000_000 / df.loc[mask, count_col]).mean()
        avg_transaction_values[count_col.replace('_count', '')] = avg_val_bdt

        service_name = count_col.replace('_count', '')
        print(f"{service_name:12}: {avg_val_bdt:8.2f} BDT per transaction")
        print(f"{'':12}  (~${avg_val_bdt/85:6.3f} USD)")  # Approximate USD conversion

# Transaction volume and value trends
print(f"\nüìä Transaction Volume vs Value Analysis:")
for i, count_col in enumerate(transaction_count_cols):
    amount_col = transaction_amount_cols[i]
    service = count_col.replace('_count', '')

    count_growth = ((df[count_col].iloc[-1] / df[count_col].iloc[0]) - 1) * 100
    amount_growth = ((df[amount_col].iloc[-1] / df[amount_col].iloc[0]) - 1) * 100

    print(f"\n{service}:")
    print(f"  Volume Growth: {count_growth:6.1f}%")
    print(f"  Value Growth:  {amount_growth:6.1f}%")
    print(f"  Growth Ratio:  {amount_growth/count_growth:.2f}" if count_growth != 0 else "  Growth Ratio: N/A")

# ================================
# PHASE 5: CORRELATION ANALYSIS
# ================================

print(f"\nüîó CORRELATION ANALYSIS")
print("=" * 40)

# Select numeric columns for correlation (before creating derived features)
numeric_cols_orig = df.select_dtypes(include=[np.number]).columns.tolist()
exclude_cols = ['Sl', 'Year', 'Month', 'Quarter']
for col in exclude_cols:
    if col in numeric_cols_orig:
        numeric_cols_orig.remove(col)

# Correlation matrix
correlation_matrix = df[numeric_cols_orig].corr()

# Plot correlation heatmap
plt.figure(figsize=(18, 14))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdYlBu_r', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .8}, fmt='.2f',
            annot_kws={'size': 8})
plt.title('Correlation Matrix of MFS Variables', fontsize=16)
plt.tight_layout()
plt.show()

# High correlations with target variables
print(f"\nüéØ Correlations with Key Metrics:")
for target in key_metrics:
    print(f"\nTop correlations with {target}:")
    target_corr = correlation_matrix[target].abs().sort_values(ascending=False)
    for i, (var, corr) in enumerate(target_corr.head(8).items()):
        if var != target:  # Exclude self-correlation
            print(f"  {i+1:2d}. {var:25s}: {corr:6.3f}")

# ================================
# PHASE 6: SEASONALITY ANALYSIS
# ================================

print(f"\nüìÖ SEASONALITY ANALYSIS")
print("=" * 40)

# Extract month and year
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter

# Monthly patterns
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Monthly transaction patterns
monthly_avg = df.groupby('Month')[key_metrics].mean()
monthly_avg.plot(kind='bar', ax=axes[0,0], title='Average Monthly Patterns')
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Average Value')
axes[0,0].tick_params(axis='x', rotation=0)
axes[0,0].legend()

# Quarterly patterns
quarterly_avg = df.groupby('Quarter')[key_metrics].mean()
quarterly_avg.plot(kind='bar', ax=axes[0,1], title='Average Quarterly Patterns')
axes[0,1].set_xlabel('Quarter')
axes[0,1].set_ylabel('Average Value')
axes[0,1].tick_params(axis='x', rotation=0)
axes[0,1].legend()

# Year-over-year growth
yearly_avg = df.groupby('Year')[key_metrics].mean()
yearly_growth = yearly_avg.pct_change() * 100
yearly_growth.plot(kind='bar', ax=axes[1,0], title='Year-over-Year Growth (%)')
axes[1,0].set_xlabel('Year')
axes[1,0].set_ylabel('Growth Rate (%)')
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].axhline(y=0, color='black', linestyle='--', alpha=0.5)
axes[1,0].legend()

# Box plot for monthly distribution
df.boxplot(column='Total_trans_count', by='Month', ax=axes[1,1])
axes[1,1].set_title('Monthly Distribution of Transaction Counts')
axes[1,1].set_xlabel('Month')
axes[1,1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))

plt.tight_layout()
plt.show()

# Seasonality insights
print(f"\nüìà Seasonal Insights:")
peak_month = monthly_avg['Total_trans_count'].idxmax()
low_month = monthly_avg['Total_trans_count'].idxmin()
peak_quarter = quarterly_avg['Total_trans_count'].idxmax()

print(f"Peak transaction month: {peak_month} ({calendar.month_name[peak_month] if 'calendar' in locals() else peak_month})")
print(f"Lowest transaction month: {low_month} ({calendar.month_name[low_month] if 'calendar' in locals() else low_month})")
print(f"Peak transaction quarter: Q{peak_quarter}")

# ================================
# PHASE 7: OUTLIER ANALYSIS
# ================================

print(f"\nüîç OUTLIER ANALYSIS")
print("=" * 40)

# Z-score based outlier detection
z_scores = np.abs(stats.zscore(df[key_metrics]))
outliers = (z_scores > 3).any(axis=1)

print(f"Number of outlier rows (Z-score > 3): {outliers.sum()}")
if outliers.sum() > 0:
    print("\nOutlier months:")
    outlier_data = df.loc[outliers, ['Months'] + key_metrics].copy()
    print(outlier_data.to_string())

    # Analyze outlier patterns
    print(f"\nOutlier Analysis:")
    for metric in key_metrics:
        outlier_vals = df.loc[outliers, metric]
        if len(outlier_vals) > 0:
            print(f"{metric}: {outlier_vals.min():,.0f} to {outlier_vals.max():,.0f}")
else:
    print("‚úÖ No significant outliers detected")

# Box plots for key metrics
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for i, metric in enumerate(key_metrics):
    box_data = axes[i].boxplot(df[metric], patch_artist=True)
    axes[i].set_title(f'{metric} Distribution')
    axes[i].set_ylabel('Value')

    # Color the box
    box_data['boxes'][0].set_facecolor('lightblue')
    box_data['boxes'][0].set_alpha(0.7)

    # Format y-axis based on metric
    if 'count' in metric.lower():
        axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))
    else:
        axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

plt.tight_layout()
plt.show()

# ================================
# PHASE 8: FEATURE ENGINEERING INSIGHTS
# ================================

print(f"\nüîß FEATURE ENGINEERING INSIGHTS")
print("=" * 40)

# Create derived features
df['CashFlow_Ratio'] = df['CashIn_amount'] / df['CashOut_amount']
df['P2P_Dominance'] = df['P2P_amount'] / df['Total_amount']
df['Transaction_Efficiency'] = df['Total_amount'] / df['Total_trans_count']
df['Float_Utilization'] = df['Total_amount'] / df['Float_amount']

# Service penetration ratios
df['CashIn_Share'] = df['CashIn_count'] / df['Total_trans_count']
df['CashOut_Share'] = df['CashOut_count'] / df['Total_trans_count']
df['P2P_Share'] = df['P2P_count'] / df['Total_trans_count']

# Rolling averages (3-month and 6-month)
windows = [3, 6]
for window in windows:
    for metric in key_metrics:
        df[f'{metric}_{window}m_avg'] = df[metric].rolling(window=window).mean()
        df[f'{metric}_growth_{window}m'] = df[metric].pct_change(periods=window) * 100

# Lag features
lags = [1, 3, 6, 12]
for lag in lags:
    for metric in key_metrics:
        df[f'{metric}_lag{lag}'] = df[metric].shift(lag)

# Trend features
for metric in key_metrics:
    df[f'{metric}_trend'] = df[metric].rolling(window=6).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0] if len(x) == 6 else np.nan
    )

print("‚úÖ Created derived features:")
derived_features = ['CashFlow_Ratio', 'P2P_Dominance', 'Transaction_Efficiency', 'Float_Utilization']
print(f"\nüìä Key Derived Features Summary:")
for feature in derived_features:
    if df[feature].notna().sum() > 0:
        print(f"  {feature:25s}: Mean={df[feature].mean():.4f}, Std={df[feature].std():.4f}")

# Visualize key derived features
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
derived_plot_features = ['Transaction_Efficiency', 'Float_Utilization', 'P2P_Dominance', 'CashFlow_Ratio']

for i, feature in enumerate(derived_plot_features):
    row, col = i // 2, i % 2
    df[feature].plot(ax=axes[row, col], title=f'{feature} Over Time', color=f'C{i}')
    axes[row, col].set_xlabel('Month Index')
    axes[row, col].set_ylabel(feature)
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ================================
# PHASE 9: ADVANCED INSIGHTS
# ================================

print(f"\nüéØ ADVANCED INSIGHTS")
print("=" * 40)

# PCA Analysis - use original numeric columns only (no derived features with NaN)
pca_cols = [col for col in numeric_cols_orig if df[col].notna().all()]
print(f"Using {len(pca_cols)} complete columns for PCA analysis")
print(f"PCA columns: {pca_cols}")

if len(pca_cols) > 3:
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[pca_cols])

    pca = PCA()
    pca_result = pca.fit_transform(scaled_data)

    # Explained variance
    plt.figure(figsize=(15, 6))

    # Cumulative explained variance
    plt.subplot(1, 3, 1)
    cumvar = np.cumsum(pca.explained_variance_ratio_)
    plt.plot(range(1, len(cumvar) + 1), cumvar, 'bo-', linewidth=2, markersize=8)
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance Ratio')
    plt.title('PCA: Cumulative Explained Variance')
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0.95, color='red', linestyle='--', alpha=0.7, label='95% Variance')
    plt.legend()

    # Individual explained variance
    plt.subplot(1, 3, 2)
    plt.bar(range(1, min(11, len(pca.explained_variance_ratio_) + 1)),
            pca.explained_variance_ratio_[:10], alpha=0.7)
    plt.xlabel('Principal Component')
    plt.ylabel('Explained Variance Ratio')
    plt.title('Individual PC Explained Variance')
    plt.grid(True, alpha=0.3)

    # Feature importance for first PC
    plt.subplot(1, 3, 3)
    pc1_importance = abs(pca.components_[0])
    pc1_features = pd.DataFrame({
        'Feature': pca_cols,
        'Importance': pc1_importance
    }).sort_values('Importance', ascending=True)

    y_pos = range(len(pc1_features))
    plt.barh(y_pos, pc1_features['Importance'], alpha=0.7)
    plt.yticks(y_pos, pc1_features['Feature'])
    plt.xlabel('Absolute Loading')
    plt.title('First PC - Feature Loadings')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print(f"\nüìä PCA Results:")
    print(f"  ‚Ä¢ First 5 PCs explain: {cumvar[4]:.1%} of variance")
    print(f"  ‚Ä¢ First 10 PCs explain: {cumvar[9]:.1%} of variance" if len(cumvar) > 9 else "")
    print(f"  ‚Ä¢ Most important features in PC1: {', '.join(pc1_features.tail(3)['Feature'].tolist())}")
else:
    print("‚ö†Ô∏è Not enough complete columns for meaningful PCA analysis")

# ================================
# PHASE 10: GROWTH TREND DECOMPOSITION
# ================================

print(f"\nüìä GROWTH TREND DECOMPOSITION")
print("=" * 40)

# Calculate comprehensive growth rates
growth_metrics = {}
for metric in key_metrics:
    growth_metrics[f'{metric}_MoM'] = df[metric].pct_change() * 100
    growth_metrics[f'{metric}_3MoM'] = df[metric].pct_change(periods=3) * 100
    if len(df) >= 12:
        growth_metrics[f'{metric}_YoY'] = df[metric].pct_change(periods=12) * 100

growth_df = pd.DataFrame(growth_metrics)

# Plot comprehensive growth analysis
fig, axes = plt.subplots(3, 1, figsize=(16, 15))

# Month-over-month growth
mom_cols = [col for col in growth_df.columns if '_MoM' in col and '_3MoM' not in col]
growth_df[mom_cols].plot(ax=axes[0], title='Month-over-Month Growth Rates (%)', linewidth=2)
axes[0].axhline(y=0, color='black', linestyle='--', alpha=0.5)
axes[0].set_ylabel('Growth Rate (%)')
axes[0].legend(loc='best')
axes[0].grid(True, alpha=0.3)

# 3-Month growth
mom3_cols = [col for col in growth_df.columns if '_3MoM' in col]
if mom3_cols:
    growth_df[mom3_cols].plot(ax=axes[1], title='3-Month Growth Rates (%)', linewidth=2)
    axes[1].axhline(y=0, color='black', linestyle='--', alpha=0.5)
    axes[1].set_ylabel('Growth Rate (%)')
    axes[1].legend(loc='best')
    axes[1].grid(True, alpha=0.3)

# Year-over-year growth
yoy_cols = [col for col in growth_df.columns if '_YoY' in col]
if yoy_cols:
    growth_df[yoy_cols].plot(ax=axes[2], title='Year-over-Year Growth Rates (%)', linewidth=2)
    axes[2].axhline(y=0, color='black', linestyle='--', alpha=0.5)
    axes[2].set_ylabel('Growth Rate (%)')
    axes[2].set_xlabel('Time Period')
    axes[2].legend(loc='best')
    axes[2].grid(True, alpha=0.3)
else:
    axes[2].text(0.5, 0.5, 'Insufficient data for YoY analysis\n(Need at least 12 months)',
                ha='center', va='center', transform=axes[2].transAxes, fontsize=14)
    axes[2].set_title('Year-over-Year Growth Rates (%)')

plt.tight_layout()
plt.show()

# Growth statistics
print(f"\nüìà Growth Rate Statistics:")
growth_stats = growth_df.describe()
print(growth_stats.round(2))

# ================================
# PHASE 11: BUSINESS INSIGHTS
# ================================

print(f"\nüéØ BUSINESS INSIGHTS")
print("=" * 40)

# Market dynamics
print(f"üìä Market Dynamics Analysis:")
latest_data = df.iloc[-1]
earliest_data = df.iloc[0]

print(f"\nüíº Market Scale (Latest Month):")
print(f"  ‚Ä¢ Total Transactions: {latest_data['Total_trans_count']:,.0f}")
print(f"  ‚Ä¢ Total Volume: {latest_data['Total_amount']:,.0f} Million BDT")
print(f"  ‚Ä¢ Average Transaction: {(latest_data['Total_amount'] * 1_000_000 / latest_data['Total_trans_count']):.2f} BDT")
print(f"  ‚Ä¢ Daily Transactions: {(latest_data['Total_trans_count'] / 30):,.0f}")

print(f"\nüìà Service Evolution:")
for i, service in enumerate(['CashIn', 'CashOut', 'P2P', 'MP', 'G2P']):
    count_col = f'{service}_count'
    amount_col = f'{service}_amount'

    if count_col in df.columns and amount_col in df.columns:
        latest_share_count = (latest_data[count_col] / latest_data['Total_trans_count']) * 100
        earliest_share_count = (earliest_data[count_col] / earliest_data['Total_trans_count']) * 100
        share_change = latest_share_count - earliest_share_count

        print(f"  ‚Ä¢ {service:8s}: {latest_share_count:5.1f}% volume share ({share_change:+5.1f}% change)")

# ================================
# PHASE 12: MODEL READINESS ASSESSMENT
# ================================

print(f"\nüéØ MODEL READINESS ASSESSMENT")
print("=" * 40)

# Feature availability for modeling
total_features = len(df.columns)
numeric_features = len(df.select_dtypes(include=[np.number]).columns)
complete_features = sum(df[col].notna().all() for col in df.select_dtypes(include=[np.number]).columns)

print(f"üìä Feature Inventory:")
print(f"  ‚Ä¢ Total Features: {total_features}")
print(f"  ‚Ä¢ Numeric Features: {numeric_features}")
print(f"  ‚Ä¢ Complete Features (no NaN): {complete_features}")
print(f"  ‚Ä¢ Features with some NaN: {numeric_features - complete_features}")

# Time series characteristics
print(f"\nüìà Time Series Characteristics:")
print(f"  ‚Ä¢ Time Series Length: {len(df)} months")
print(f"  ‚Ä¢ Date Range: {df['Date'].min().strftime('%B %Y')} to {df['Date'].max().strftime('%B %Y')}")
print(f"  ‚Ä¢ Frequency: Monthly")
print(f"  ‚Ä¢ Suitable for ML: {'‚úÖ Yes' if len(df) >= 24 else '‚ö†Ô∏è Limited (recommended: 24+ months)'}")

# ================================
# PHASE 13: COMPREHENSIVE SUMMARY
# ================================

print(f"\nüìã COMPREHENSIVE EDA SUMMARY")
print("=" * 60)

print("üöÄ KEY GROWTH TRENDS:")
total_trans_growth = ((df['Total_trans_count'].iloc[-1] / df['Total_trans_count'].iloc[0]) - 1) * 100
total_amount_growth = ((df['Total_amount'].iloc[-1] / df['Total_amount'].iloc[0]) - 1) * 100
float_growth = ((df['Float_amount'].iloc[-1] / df['Float_amount'].iloc[0]) - 1) * 100

print(f"  ‚Ä¢ Transaction Volume: {total_trans_growth:.0f}% total growth")
print(f"  ‚Ä¢ Transaction Value: {total_amount_growth:.0f}% total growth")
print(f"  ‚Ä¢ Float Amount: {float_growth:.0f}% total growth")
print(f"  ‚Ä¢ Average Transaction Size: {((latest_data['Total_amount'] * 1_000_000 / latest_data['Total_trans_count']) / (earliest_data['Total_amount'] * 1_000_000 / earliest_data['Total_trans_count']) - 1) * 100:.1f}% change")

print(f"\nüí≥ TRANSACTION ECOSYSTEM:")
dominant_service = df[transaction_count_cols].iloc[-1].idxmax().replace('_count', '')
print(f"  ‚Ä¢ Dominant Service: {dominant_service} ({(df[transaction_count_cols].iloc[-1].max() / df['Total_trans_count'].iloc[-1] * 100):.1f}% of volume)")
print(f"  ‚Ä¢ Service Diversity: {len([col for col in transaction_count_cols if df[col].iloc[-1] > 0])} active transaction types")
print(f"  ‚Ä¢ Cash-dominated: {'Yes' if df['CashIn_count'].iloc[-1] + df['CashOut_count'].iloc[-1] > df['Total_trans_count'].iloc[-1] * 0.6 else 'No'}")

print(f"\nüîó KEY RELATIONSHIPS:")
# Get strongest correlations for each key metric
strong_correlations = []
for target in key_metrics:
    correlations = correlation_matrix[target].abs().sort_values(ascending=False)
    top_corr_var = correlations.index[1]  # Exclude self-correlation
    top_corr_val = correlations.iloc[1]
    strong_correlations.append(f"  ‚Ä¢ {target} ‚Üî {top_corr_var}: {top_corr_val:.3f}")

for corr in strong_correlations:
    print(corr)

print(f"\nüìä DATA QUALITY:")
data_completeness = ((df.shape[0] * numeric_features - df.select_dtypes(include=[np.number]).isnull().sum().sum()) / (df.shape[0] * numeric_features) * 100)
print(f"  ‚Ä¢ Dataset Completeness: {data_completeness:.1f}%")
print(f"  ‚Ä¢ Time Series Continuity: {'‚úÖ Complete' if df['Date'].diff().dropna().nunique() == 1 else '‚ö†Ô∏è Gaps detected'}")
print(f"  ‚Ä¢ Outlier Prevalence: {outliers.sum()}/{len(df)} months ({(outliers.sum()/len(df)*100):.1f}%)")

print(f"\nüéØ MODELING RECOMMENDATIONS:")
print("  ‚úÖ FEATURES TO USE:")
print("    ‚Ä¢ Lagged features (1, 3, 6, 12 months) for time series prediction")
print("    ‚Ä¢ Rolling averages (3, 6 months) to smooth volatility")
print("    ‚Ä¢ Derived ratios (Transaction_Efficiency, Float_Utilization)")
print("    ‚Ä¢ Seasonal features (Month, Quarter)")
print("    ‚Ä¢ Growth momentum features")

print(f"\n  üìà MODEL STRATEGY:")
print("    ‚Ä¢ Linear Models: Good baseline with interpretable coefficients")
print("    ‚Ä¢ Tree-based: Excellent for capturing non-linear patterns")
print("    ‚Ä¢ XGBoost/LightGBM: Expected top performers for tabular data")
print("    ‚Ä¢ LSTM: Ideal for capturing temporal dependencies")
print("    ‚Ä¢ Ensemble: Combine multiple models for robust predictions")

print(f"\n  üé≤ TARGET VARIABLE RECOMMENDATIONS:")
print("    ‚Ä¢ Primary: Total_trans_count (highest business impact)")
print("    ‚Ä¢ Secondary: Float_amount (liquidity management)")
print("    ‚Ä¢ Tertiary: Service-specific volumes for strategic planning")

print(f"\n  ‚öôÔ∏è PREPROCESSING STEPS:")
print("    ‚Ä¢ Standard scaling for linear models and neural networks")
print("    ‚Ä¢ Handle seasonality with cyclical encoding")
print("    ‚Ä¢ Create interaction features between services")
print("    ‚Ä¢ Use time-based train/validation splits")

print(f"\nüéä BUSINESS INSIGHTS:")
print(f"  ‚Ä¢ Market Maturity: {'High' if total_trans_growth > 200 else 'Medium' if total_trans_growth > 100 else 'Early'} (based on {total_trans_growth:.0f}% growth)")
print(f"  ‚Ä¢ Transaction Pattern: High-frequency, low-value (avg: {(latest_data['Total_amount'] * 1_000_000 / latest_data['Total_trans_count']):.2f} BDT)")
print(f"  ‚Ä¢ Growth Driver: {'Volume-driven' if total_trans_growth > total_amount_growth else 'Value-driven'} expansion")
print(f"  ‚Ä¢ Seasonality: {'Present' if monthly_avg['Total_trans_count'].std() / monthly_avg['Total_trans_count'].mean() > 0.05 else 'Minimal'}")

print(f"\nüîÆ PREDICTIVE MODELING OUTLOOK:")
suitable_models = []
if len(df) >= 36:
    suitable_models.extend(["LSTM", "Advanced Time Series"])
if numeric_features >= 15:
    suitable_models.extend(["Random Forest", "XGBoost", "LightGBM"])
suitable_models.extend(["Linear Regression", "Ridge", "ElasticNet"])

print(f"  ‚Ä¢ Recommended Models: {', '.join(suitable_models)}")
print(f"  ‚Ä¢ Expected Performance: {'High' if len(df) >= 36 and numeric_features >= 15 else 'Good'}")
print(f"  ‚Ä¢ Interpretability: Mixed (linear models: high, tree-based: medium, LSTM: low)")

print(f"\n‚úÖ EDA COMPLETE!")
print("=" * 60)
print("üöÄ Your dataset is ready for comprehensive machine learning modeling!")
print(f"üìä Key Statistics: {len(df)} months, {numeric_features} features, {total_trans_growth:.0f}% growth")
print("üéØ Next Step: Proceed with the 11-model machine learning pipeline")
print("=" * 60)

# Save processed dataset for modeling
import calendar
df.to_csv('mfs_processed_for_modeling.csv', index=False)
print("üíæ Processed dataset saved as: 'mfs_processed_for_modeling.csv'")
print("üìÅ This file contains all original features plus engineered features for modeling")
```
